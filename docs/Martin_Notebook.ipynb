```python
import nltk
#imports
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer
import ast
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
import stanza
import nltk
```

```python
nltk.download('stopwords')
german_stop_words = nltk.corpus.stopwords.words('german')
stanza.download('de')
```

```python
# changed to relative path
WD='../'
data_path=WD+'data/input/'
```

separate to labels and features

```python
train_df=pd.read_csv(data_path+'train_data.csv')
valid_df=pd.read_csv(data_path+'valid_data.csv')
```

```python
train_features =   train_df.iloc[:,[0,1]].copy()
train_labels =  train_df.iloc[:,[0,2]].copy()
```

```python
nlp=stanza.Pipeline(processors="tokenize,mwt,lemma", lang="de")
def tokenize(input: str):
    doc=nlp(input)
    return[word.lemma  for sent in doc.sentences for word in sent.words if word.lemma not in german_stop_words]
```

```python
train_features['tokens']=train_features.Text.apply(tokenize)
```

```python
train_features
```

```python
vectorizer=CountVectorizer(tokenizer=tokenize)
```

```python
Vec=vectorizer.fit_transform(train_features.Text)
```

```python
Vec=Vec.toarray()
```

```python
train_features
```

```python
## get rid of numbers? They can have some informational value
vectorizer.get_feature_names_out()
```

```python
vectorizer.get_feature_names_out
```

```python
train_features[vectorizer.get_feature_names_out()]=Vec
```

```python
train_features
```

```python
train_labels.Labels=train_labels.Labels.apply(ast.literal_eval)
mlb=MultiLabelBinarizer()
train_labels_transformed=mlb.fit_transform(train_labels['Labels'])
train_labels[mlb.classes_]=train_labels_transformed
```

```python
train_labels
```

```python
train_features.to_csv((data_path+'features.csv'), index=False)
train_labels.to_csv((data_path+'labels.csv'), index=False)
```
